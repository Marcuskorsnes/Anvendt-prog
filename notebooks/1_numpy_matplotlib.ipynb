### b) Visualisere innholdet i Medals.xlsx

I denne oppgaven skal du visualisere innholdet i filen ved hjelp av Python-biblioteker som Matplotlib.
Bruk Matplotlib til å lage grafer som viser fordelingen av medaljer.    


```python
import pandas as pd
import matplotlib.pyplot as plt

medals_data = pd.read_excel("data/Medals.xlsx")

print(medals_data.columns)


countries = medals_data["Team/NOC"]
gold_medals = medals_data["Gold"]
silver_medals = medals_data["Silver"]
bronze_medals = medals_data["Bronze"]

# Velger topp 10 land for bedre oversikt
top_10_gold = medals_data[['Team/NOC', 'Gold']].sort_values(by='Gold', ascending=False).head(10)
top_10_silver = medals_data[['Team/NOC', 'Silver']].sort_values(by='Silver', ascending=False).head(10)  
top_10_bronze = medals_data[['Team/NOC', 'Bronze']].sort_values(by='Bronze', ascending=False).head(10)


fig, ax = plt.subplots(2, 2, figsize=(14, 10))


ax[0, 0].bar(top_10_gold['Team/NOC'], top_10_gold['Gold'], color="gold")
ax[0, 0].set_title("Topp 10 Gullmedaljer per Land")
ax[0, 0].set_xlabel("Land")
ax[0, 0].set_ylabel("Antall Gullmedaljer")
ax[0, 0].tick_params(axis='x', rotation=45) 
ax[0, 0].set_xticks(range(len(top_10_gold)))  
ax[0, 0].set_xticklabels(top_10_gold['Team/NOC'], rotation=45, ha="right")

ax[0, 1].bar(top_10_silver['Team/NOC'], top_10_silver['Silver'], color="silver")
ax[0, 1].set_title("Topp 10 Sølvmedaljer per Land")
ax[0, 1].set_xlabel("Land")
ax[0, 1].set_ylabel("Antall Sølvmedaljer")
ax[0, 1].tick_params(axis='x', rotation=45)
ax[0, 1].set_xticks(range(len(top_10_silver))) 
ax[0, 1].set_xticklabels(top_10_silver['Team/NOC'], rotation=45, ha="right")

ax[1, 0].bar(top_10_bronze['Team/NOC'], top_10_bronze['Bronze'], color="brown")
ax[1, 0].set_title("Topp 10 Bronsemedaljer per Land")
ax[1, 0].set_xlabel("Land")
ax[1, 0].set_ylabel("Antall Bronsemedaljer")
ax[1, 0].tick_params(axis='x', rotation=45)
ax[1, 0].set_xticks(range(len(top_10_bronze))) 
ax[1, 0].set_xticklabels(top_10_bronze['Team/NOC'], rotation=45, ha="right")


top_10_total = medals_data[['Team/NOC', 'Gold', 'Silver', 'Bronze']].sort_values(by=['Gold', 'Silver', 'Bronze'], ascending=False).head(10)
top_10_total['Total'] = top_10_total['Gold'] + top_10_total['Silver'] + top_10_total['Bronze']

ax[1, 1].bar(top_10_total['Team/NOC'], top_10_total['Total'], color="royalblue")
ax[1, 1].set_title("Topp 10 Totalt Antall Medaljer per Land")
ax[1, 1].set_xlabel("Land")
ax[1, 1].set_ylabel("Totalt Antall Medaljer")
ax[1, 1].tick_params(axis='x', rotation=45)
ax[1, 1].set_xticks(range(len(top_10_total)))  
ax[1, 1].set_xticklabels(top_10_total['Team/NOC'], rotation=45, ha="right")


plt.tight_layout()


plt.subplots_adjust(hspace=0.4, wspace=0.4)


plt.show()

```

### c) Visualisere innholdet i EntriesGender.xlsx

I denne oppgaven skal du åpne filen EntriesGender.xlsx, som inneholder informasjon om kjønnsfordelingen av deltakere i forskjellige arrangementer under de olympiske leker i Tokyo 2021. Du skal deretter visualisere innholdet i filen ved hjelp av Python-biblioteker som Pandas og Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt


entries_gender_data = pd.read_excel("data/EntriesGender.xlsx")

print(entries_gender_data.columns)


events = entries_gender_data['Discipline']  
male_count = entries_gender_data['Male']  
female_count = entries_gender_data['Female'] 


fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = range(len(events))


ax.bar(index, male_count, bar_width, label='Male', color='blue')
ax.bar([i + bar_width for i in index], female_count, bar_width, label='Female', color='pink')


ax.set_xlabel('Discipline')
ax.set_ylabel('Number of Participants')
ax.set_title('Gender Distribution of Participants in Olympic Disciplines (Tokyo 2021)')


ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(events, rotation=90)


ax.legend()


plt.tight_layout()
plt.show()

```

### d) Analyser medaljefordelingen etter kjønn (Vanskelig)

I denne oppgaven (frivillig) skal du kombinere data fra Medals.xlsx og EntriesGender.xlsx for å analysere medaljefordelingen etter kjønn. Du skal deretter visualisere resultatene ved hjelp av Python-biblioteker som Pandas og Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt


medals_data = pd.read_excel("data/Medals.xlsx")
entries_gender_data = pd.read_excel("data/EntriesGender.xlsx")


medals_data['Female_Gold'] = medals_data['Gold'] * (entries_gender_data['Female'] / entries_gender_data['Total'])
medals_data['Male_Gold'] = medals_data['Gold'] * (entries_gender_data['Male'] / entries_gender_data['Total'])
medals_data['Female_Silver'] = medals_data['Silver'] * (entries_gender_data['Female'] / entries_gender_data['Total'])
medals_data['Male_Silver'] = medals_data['Silver'] * (entries_gender_data['Male'] / entries_gender_data['Total'])
medals_data['Female_Bronze'] = medals_data['Bronze'] * (entries_gender_data['Female'] / entries_gender_data['Total'])
medals_data['Male_Bronze'] = medals_data['Bronze'] * (entries_gender_data['Male'] / entries_gender_data['Total'])


fig, ax = plt.subplots(figsize=(12, 8))

ax.bar(medals_data['Team/NOC'], medals_data['Female_Gold'], label='Female Gold', color='orange')
ax.bar(medals_data['Team/NOC'], medals_data['Male_Gold'], label='Male Gold', color='gold', bottom=medals_data['Female_Gold'])

ax.bar(medals_data['Team/NOC'], medals_data['Female_Silver'], label='Female Silver', color='darkgrey', bottom=medals_data['Female_Gold'] + medals_data['Male_Gold'])
ax.bar(medals_data['Team/NOC'], medals_data['Male_Silver'], label='Male Silver', color='silver', bottom=medals_data['Female_Gold'] + medals_data['Male_Gold'] + medals_data['Female_Silver'])

ax.bar(medals_data['Team/NOC'], medals_data['Female_Bronze'], label='Female Bronze', color='brown', bottom=medals_data['Female_Gold'] + medals_data['Male_Gold'] + medals_data['Female_Silver'] + medals_data['Male_Silver'])
ax.bar(medals_data['Team/NOC'], medals_data['Male_Bronze'], label='Male Bronze', color='red', bottom=medals_data['Female_Gold'] + medals_data['Male_Gold'] + medals_data['Female_Silver'] + medals_data['Male_Silver'] + medals_data['Female_Bronze'])


ax.set_xlabel('Team/NOC')
ax.set_ylabel('Number of Medals')
ax.set_title('Total Medal Distribution by Gender for Each Country (Tokyo 2021)')
ax.set_xticklabels(medals_data['Team/NOC'], rotation=90)


ax.legend()

plt.tight_layout()
plt.show()
```
