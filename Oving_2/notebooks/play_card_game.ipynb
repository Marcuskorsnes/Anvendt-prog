````python
def meny():
    while True:
        print("\nVelkommen til kortspillet!\n")
        print("Del hånd")
        print("Vis hånd")
        print("Sjekk hånd")
        print("Avslutt\n")

        valg = input("Hva vil du gjøre? ").strip().lower()

        print()

        if valg == "del hånd":
            print("Deler ut ny hånd med 5 kort\n")
        elif valg == "vis hånd":
            print("Viser hånd\n")
        elif valg == "sjekk hånd":
            print("Sjekker hånd\n")
        elif valg == "avslutt":
            print("Avslutter spillet\n")
            break
        else:
            print("Ugyldig valg\n")

meny()
````

### Oppgave 2

I denne oppgaven skal du analysere kortene på hånd (se punktliste under). Dette kan skje når brukeren trykker på menyvalget "Check hand".

Her er det meningen at du skal utvide metoden check_hand samt å anvende streams med tilhørende funskjoner for filtrering, map, reduce osv. Prøv å løse så mange som du klarer av følgende:

- Regn ut summen av alle verdiene av kortene på hånd (ess = 1) 
- Hent ut bare kort som er av fargen "Hjerter", og vis i et tekstfelt på formen "H12 H9 H1". Dersom det ikke er noen Hjerter på hånd, kan tekstfeltet inneholde teksten "No Hearts", for eksempel.
- Sjekk om kortet "Spar dame" finnes blant kortene på hånden.


````python
from functools import reduce


class PlayingCard:
    def __init__(self, type, verdi):
        self.type = type  # Farge på kortet
        self.verdi = verdi  # Verdi på kortet 
    
    def get_type(self):
        return self.type  
    
    def get_verdi(self):
        return self.verdi 
    
    def get_as_string(self):
        return f"{self.type}{self.verdi}"  # Returnerer kortet som en streng (f.eks. H12 for Hjerter dame)


class HandOfCards:
    def __init__(self, cards):
        self.cards = cards  
    
    def get_cards(self):
        return self.cards  # Returnerer listen av kortene på hånden

# Funksjon for å sjekke hånd
def check_hand(hand):
    """Analyserer kortene på hånden."""
    cards = hand.get_cards()  # Henter alle kortene fra hånden

 
    total_value = reduce(lambda acc, card: acc + (1 if card.get_verdi() == 1 else card.get_verdi()), cards, 0)

    # Filtrerer ut kort med fargen "Hjerter"
    hearts = filter(lambda card: card.get_type() == 'H', cards)
    heart_str = " ".join(map(lambda card: card.get_as_string(), hearts)) or "No Hearts"

  
    has_queen_of_spades = any(card.get_type() == 'S' and card.get_verdi() == 12 for card in cards)

 
    print(f"\nTotal verdi av kortene: {total_value}")
    print(f"Hjerter på hånd: {heart_str}")
    print("Spar dame er på hånden" if has_queen_of_spades else "Spar dame er ikke på hånden")
    print()


def meny():
    """Håndterer menyen for kortspillet."""
    # Eksemplarisk hånd med kort
    hand = HandOfCards([
        PlayingCard('H', 12),  # Hjerter dame
        PlayingCard('H', 9),   # Hjerter 9
        PlayingCard('S', 12),  # Spar dame
        PlayingCard('D', 7),   # Diamant 7
        PlayingCard('H', 1)    # Hjerter ess
    ])

    while True:
        print("\nVelkommen til kortspillet!\n")
        print("1. Vis hånd")
        print("2. Sjekk hånd")
        print("3. Avslutt\n")

        valg = input("Hva vil du gjøre? ").strip().lower()

        if valg == "1" or valg == "vis hånd":
            print("Hånd:", ", ".join(card.get_as_string() for card in hand.get_cards()), "\n")  # Henter kortene
        elif valg == "2" or valg == "sjekk hånd":
            check_hand(hand)
        elif valg == "3" or valg == "avslutt":
            print("Avslutter spillet.\n")
            break
        else:
            print("Ugyldig valg, prøv igjen.\n")

meny()
````

````python
import unittest
import sys, os

# Add the parent directory of 'notebooks' to the sys.path
sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), 'src')))

from DeckOfCards import DeckOfCards


class TestDeckOfCards(unittest.TestCase):

    def test_deck_creation(self):
        deck = DeckOfCards()
        self.assertEqual(len(deck.cards), 52, "Deck should have 52 cards.")
    
    def test_deal_hand(self):
        deck = DeckOfCards()
        hand = deck.deal_hand(5)
        self.assertEqual(len(hand.get_cards()), 5, "Hand should have 5 cards.")
    
    def test_shuffle(self):
        deck = DeckOfCards()
        initial_cards = deck.cards[:]
        deck.shuffle()
        self.assertNotEqual(initial_cards, deck.cards, "Deck should be shuffled.")

if __name__ == "__main__":
    unittest.main()

````
